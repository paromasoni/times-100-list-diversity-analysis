```python
def group_num(year):
    if year == 2004:
        return '1'
    
    if year == 2007:
        return '2'

    if year == 2010:
        return '3'
    
    if year == 2013:
        return '4'

    if year == 2015:
        return '5'

    if year == 2017:
        return '6'

    
    if year == 2019:
        return '7'

    if year == 2020:
        return '8'
    
    else:
        return year
    
```

```python
def group_name(num):
    if num == '1':
        return '2004'
    
    if num == '2':
        return '2007'

    
    if num == '3':
        return '2010'

    
    if num == '4':
        return '2013'

    
    if num == '5':
        return '2015'

    
    if num == '6':
        return '2017'

    
    if num == '7':
        return '2019'

    
    if num == '8':
        return '2020'
    else:
        return num
```

```python
mapdf['properties.group_id'] = mapdf['properties.year'].apply(group_num)
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```
